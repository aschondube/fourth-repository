```python
df_Full = pd.read_csv('CBS_Full.csv') 
```

```python
df_sankey = df_Full[['Start Station', 'End Station']]
```

```python
df_sankey.shape
```

```python
df_sankey.iloc[1].to_dict()
```

```python
import plotly.graph_objects as go
import plotly.offline as pyo
import pandas as pd

# Group by 'Start Station' and 'End Station' and count the number of trips
sankey_data = df_sankey.groupby(['Start Station', 'End Station']).size().reset_index(name='Count')

# Create nodes from unique stations
nodes = pd.Index(pd.concat([sankey_data['Start Station'], sankey_data['End Station']]).unique())

# Create dictionary to map stations to indices
station_to_index = {station: idx for idx, station in enumerate(nodes)}

# Create source and target indices for Sankey diagram
source_indices = sankey_data['Start Station'].map(station_to_index)
target_indices = sankey_data['End Station'].map(station_to_index)

# Create Sankey diagram
fig = go.Figure(data=[go.Sankey(
    node=dict(
        pad=15,
        thickness=20,
        line=dict(color="black", width=0.5),
        label=nodes,
        color="blue"
    ),
    link=dict(
        source=source_indices,
        target=target_indices,
        value=sankey_data['Count'],
        color="blue"
    )
)])

fig.update_layout(title_text="Bike Trips Sankey Diagram")

# Display plot using Plotly Offline
pyo.plot(fig, filename='sankey_diagram.html', auto_open=True)
```
